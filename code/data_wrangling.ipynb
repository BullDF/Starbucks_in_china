```python
import pandas as pd
original_data_dir = '../original_data/'
clean_data_dir = '../clean_data/'
```

```python
starbucks = pd.read_csv(original_data_dir + 'starbucks.csv')
gdp = pd.read_csv(original_data_dir + 'gdp.csv')
shanghai = pd.read_csv(original_data_dir + 'shanghai_2019.csv')
population = pd.read_csv(original_data_dir + 'population.csv')
```

# Shanghai

```python
districts = [
    "Pudong New Area",
    "Huangpu",
    "Xuhui",
    "Changning",
    "Jing'an",
    "Putuo",
    "Hongkou",
    "Yangpu",
    "Minhang",
    "Baoshan",
    "Jiading",
    "Jinshan",
    "Songjiang",
    "Qingpu",
    "Fengxian",
    "Chongming"
]
```

```python
shanghai
```

```python
shanghai_en = shanghai.rename(columns={'地区': 'District',
                                       '行政区划面积（平方公里）': 'Area',
                                       '年末常住人口（万人）': 'Population',
                                       '人口密度（人/平方公里）': 'Density'})
shanghai_en = shanghai_en.drop('其中外来人口', axis=1)
shanghai_en['District'] = districts
shanghai_en
```

```python
shanghai_en.to_csv(clean_data_dir + 'shanghai_2019_en.csv', index=False)
```

```python
population_clean = population[~population['Province'].isin(['Servicemen', 'Hainan'])]
population_clean
```

```python
gdp.columns[1:].tolist()
```

```python
population_clean['Province'].tolist()
```

```python
gdp_clean = gdp.rename(columns={'Guangxi,': 'Guangxi'})
gdp_clean = gdp_clean.drop('Hainan', axis=1)
gdp_clean
```

```python
set(gdp_clean.columns[1:]).difference(set(population_clean['Province']))
```

```python
set(population_clean['Province']).difference(set(gdp_clean.columns[1:]))
```

```python
population_clean
```

```python
population_clean.to_csv(clean_data_dir + 'population.csv', index=False)
```

```python
gdp_final = gdp_clean[gdp_clean['Year'] == 2019]
gdp_final = gdp_final.drop('Year', axis=1)
gdp_final = gdp_final.T.reset_index()
gdp_final = gdp_final.rename(columns={'index': 'Province', 1: 'GDP'})
gdp_final
```

```python
gdp_final.to_csv(clean_data_dir + 'gdp_2019.csv', index=False)
```
